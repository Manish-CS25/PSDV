```python
# Q1 List Operations and Basic Statistics:

# 1. Create a list of the first 20 natural numbers.
natural_numbers = list(range(1, 21))
print(f"First 20 natural numbers: {natural_numbers}")

# 2. Compute the sum of all even numbers in the list.
even_sum = sum([num for num in natural_numbers if num % 2 == 0])
print(f"Sum of even numbers: {even_sum}")

# 3. Compute the product of all odd numbers in the list.
odd_product = 1
for num in natural_numbers:
    if num % 2 != 0:
        odd_product *= num
print(f"Product of odd numbers: {odd_product}")

# 4. Calculate the mean, median, and standard deviation of the list.

mean = sum(natural_numbers) / len(natural_numbers)
sorted_numbers = sorted(natural_numbers)
middle = len(sorted_numbers) // 2
median = (sorted_numbers[middle] + sorted_numbers[~middle]) / 2
std_dev = (sum((x - mean) ** 2 for x in natural_numbers) / len(natural_numbers)) ** 0.5

print(f"Mean: {mean}, Median: {median}, Standard Deviation: {std_dev}")
```

```python
# Q3.Fibonacci Sequence:

# 1. Write a Python function that takes a number n and returns a list of the first n Fibonacci numbers
def fibonacci(n):
    fibno_list = [0, 1]
    for i in range(2, n):
        fibno_list.append(fibno_list[i - 1] + fibno_list[i - 2])
    return fibno_list

# 2.Plot the Fibonacci sequence using a line graph.
import matplotlib.pyplot as plt
print(f"list of first 10 fibonacci numbers: {fibonacci(10)}")
plt.plot(fibonacci(10))
plt.show()
```

```python
#Q6. Palindrome Check:
# Write a program to check if a string is a palindrome (ignoring spaces and case).
# Input: "A man a plan a canal Panama"
# Output: True

def is_palindrome(text):
  text = text.replace(" ", "").lower()
  return text == text[::-1]

user_input = input("Enter a string: ")
if is_palindrome(user_input):
  print("True")
else:
  print("False")
```

```python
# Q9. Merging Dictionaries:
# Write Python code to merge two dictionaries. If a key is present in both, sum their values.
# Example:
# dict1 = {'a': 10, 'b': 20}
# dict2 = {'b': 5, 'c': 15}
# Output: {'a': 10, 'b': 25, 'c': 15}

dict1 = {'a': 10, 'b': 20}
dict2 = {'b': 5, 'c': 15}
for key, value in dict2.items():
  if key in dict1:
    dict1[key] += value
  else:
    dict1[key] = value
print(f"Merged dictionary: {dict1}")

```

```python
# Q 10. Character Frequency:
# Write a function that accepts a string and:
def char_freq(string):
  # Converts all characters to lowercase.
  string.lower()
  freq = {}
  # Counts the frequency of each character.
  for char in string:
    if char in freq:
      freq[char] += 1
    else:
      freq[char] = 1
  # Returns a dictionary with characters as keys and their counts as values.
  print(f"Character frequency: {freq}")
  return freq


# Visualize the character frequencies using a bar chart
import matplotlib.pyplot as plt
plt.bar(char_freq("hello").keys(), char_freq("hello").values())
plt.show()
```

```python
#Q11. Remove Duplicates:
# Write a program to remove duplicate elements from a list while maintaining the order of elements.
list = [1, 2, 2, 3, 4, 4, 5]
unique_list = []
for element in list:
  if element not in unique_list:
    unique_list.append(element)
# Calculate the mean and median of the unique elements.
mean = sum(unique_list) / len(unique_list)
sorted_unique_list = sorted(unique_list)
middle = len(sorted_unique_list) // 2
median = (sorted_unique_list[middle] + sorted_unique_list[~middle]) / 2
print(f"Mean: {mean}, Median: {median}")
```

```python
# Q15 . Anagram Finder:
# Write a Python function to find all anagrams of a word in a given list.
# Input: word = "listen", words = ["enlist", "google", "inlets", "banana"]
# Output: ["enlist", "inlets"]

word = "listen"
words = ["enlist", "google", "inlets", "banana"]

def find_anagrams(word, words):
  anagrams = []
  for w in words:
    if sorted(word) == sorted(w):
      anagrams.append(w)
  return anagrams

print(f"Anagrams: {find_anagrams(word, words)}")
```

```python
# Q22. Normalization:
# Write a Python function to normalize a list of numbers to a range of 0 to 1.
myList = [13 , 14 , 10 , 3 , 5, 9 , 11]
def normalize(myList):
  min = myList[0]
  max = myList[0]
  for num in myList:
    if num < min:
      min = num
    if num > max:
      max = num
  for i in range(len(myList)):
    myList[i] = (myList[i] - min) / (max - min)
  return myList
print(f"Normalized list: {normalize(myList)}")
# Visualize the normalized data using a line plot.
import matplotlib.pyplot as plt
fig_width = 9
fig_height = 6
plt.figure(figsize=(fig_width, fig_height))
plt.plot(normalize(myList))
plt.show()
```

```python
# Q24. Grade Calculator:
# Write a Python program that takes a numerical grade (0-100) as input and prints the corresponding letter grade (A, B, C, D, F).
marks = int(input("Enter Your marks:-"))
def grade(marks):
  if marks >= 90:
    return "A"
  elif marks >= 80:
    return "B"
  elif marks >= 70:
    return "C"
  elif marks >= 60:
    return "D"
  else:
    return "F"

print(f"Grade: {grade(marks)}")
```
